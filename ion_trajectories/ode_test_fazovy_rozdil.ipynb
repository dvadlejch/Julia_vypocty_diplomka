## ODE Test fazovy rozdil

Zkusím řešit čistě rovnici $  \ddot{u} =  k \sin{(\Omega t)} $.

```julia
using DifferentialEquations

k = 1.0
Ω = 1.0
u0 = [-k/Ω, 0.0] # [vu, u]

function ode_fun!(du,u,p,t)
    # p=k
    du[1] = p[1] * sin(p[2]*t)
    du[2] = u[1]
end

p = [k, Ω]
tspan = (0.0, 20.0)

prob = ODEProblem(ode_fun!, u0, tspan, p)
sol = solve(prob, abstol=1e-8, reltol=1e-8)
```

```julia
#using Plots

gr()
plot(sol, vars=(0,2))
```
