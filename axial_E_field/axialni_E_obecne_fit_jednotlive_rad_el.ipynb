# Axiální složka el. pole v případě všech jednotlivých elektrod uvažovaných samostatně

nejprve naimportuju el. pole generované jednou radiální elektrodou a el. pole generované jednou axiální elektrodou na ose $z$.

```julia
# import jednotlivych osovych poli z comsolu
# 13 - par rad. el. y
# 24 - par rad. el. x
# 5 - axialni elektroda z-
# 6 - axialni elektroda z+



using DelimitedFiles

## nacteni dat ze souboru
# pole generovano axialni elektrodou v + 2.25mm
z_ax_data = readdlm("axial_el_unitary_field_delim.txt")[:,2]
E_field_ax_data = readdlm("axial_el_unitary_field_delim.txt")[:,3];

# vyberu pouze interval -1mm az 1mm
leng_ax = length(z_ax_data)
z_ax = z_ax_data[convert(Int64,round(leng_ax/4)):convert(Int64,round(3*leng_ax/4))]*1e-3
E_field_ax = E_field_ax_data[convert(Int64,round(leng_ax/4)):convert(Int64,round(3*leng_ax/4))]

# pole generovano radiální elektrodou
z_rad_data = readdlm("one_rad_el_unitary_field_delim.txt")[:,1]
E_field_rad_data = readdlm("one_rad_el_unitary_field_delim.txt")[:,2];

# vyberu pouze interval -1mm az 1mm
# dale pak vyuziju symetrie
leng_rad = length(z_rad_data)
z_rad = reverse(-z_rad_data[1:convert(Int64,round(leng_rad/2))]) * 1e-3
append!(z_rad,z_rad_data[1:convert(Int64,round(leng_rad/2))]*1e-3)

E_field_rad = reverse(-E_field_rad_data[1:convert(Int64,round(leng_rad/2))])
append!(E_field_rad, E_field_rad_data[1:convert(Int64,round(leng_rad/2))])

z_rad_cele = reverse(-z_rad_data) * 1e-3
append!(z_rad_cele,z_rad_data*1e-3)

E_field_rad_cele = reverse(-E_field_rad_data)
append!(E_field_rad_cele, E_field_rad_data);

# vyuziti symetrie

# pole generovano parem radialnich elektrod
z_rad_data_par = readdlm("rad_el_unitary_field_delim.txt")[:,1]
E_field_rad_data_par = readdlm("rad_el_unitary_field_delim.txt")[:,2];

# vyberu pouze interval -1mm az 1mm
# dale pak vyuziju symetrie
leng_rad_par = length(z_rad_data_par)
z_rad_par = reverse(-z_rad_data_par[1:convert(Int64,round(leng_rad_par/2))]) * 1e-3
append!(z_rad_par,z_rad_data_par[1:convert(Int64,round(leng_rad_par/2))]*1e-3)

E_field_rad_par = reverse(-E_field_rad_data_par[1:convert(Int64,round(leng_rad_par/2))])
append!(E_field_rad_par, E_field_rad_data_par[1:convert(Int64,round(leng_rad_par/2))])
```

```julia
# import packagu pro ploceni
using PyPlot
using PyCall
using LaTeXStrings
using Formatting

# latex fonty
rcParams = PyPlot.PyDict(PyPlot.matplotlib."rcParams")
rcParams["font.size"] = 14
rcParams["mathtext.fontset"] = "cm"
rcParams["mathtext.rm"] = "serif"
rcParams["text.usetex"] = "false";
```

```julia
my_dpi = 150
siz = 100
sizex = 16*siz
sizey = 9*siz

figure(figsize=(sizex/my_dpi, sizey/my_dpi), dpi=my_dpi)

x=gca()

line1 = x.plot(z_rad, 2*E_field_rad, color="black", label="E_rad")
#line2 = x.plot(z_ax, E_field_ax, color="blue", label="E_ax")
line2 = x.plot(z_rad_par, E_field_rad_par, color="red", label="E_rad_par")
line3 = x.plot(z_ax, E_field_ax, color="blue", label="E_ax")
#line4 = x.plot(z_ax, - reverse(E_field_ax), color="blue", label="E_ax")
x.set_xlabel(L"z \,\, \rm{[\rm{\mu} m]}", fontsize=16)
#x.set_ylabel(L"E_{\rm{rf},\it{z}} \, \, \rm{[V/m]}", fontsize=16)

x.grid()
x.legend(framealpha=1)
```

Dle očekávání je osové elektrické pole generované jednou radiální elektrodou poloviční vzhledem k poli generovaném párem protějších radiálních elektrod.
