```julia
# import jednotlivych osovych poli z comsolu
# 13 - par rad. el. y
# 24 - par rad. el. x
# 5 - axialni elektroda z-
# 6 - axialni elektroda z+

# ze symetrie: E_13 = E_24
#              E_5(z) = - E_6(-z)

using DelimitedFiles

## nacteni dat ze souboru
# pole generovano axialni elektrodou v + 2.25mm
z_data = readdlm("axial_field_unitary_rad_ax_delim.txt")[:,2]
E_field_ax_data = readdlm("axial_field_unitary_rad_ax_delim.txt")[:,3]
E_field_rad_data = readdlm("axial_field_unitary_rad_ax_delim.txt")[:,4]

leng_data = length(z_data)
# nyni mam nacteny data ve stejnych bodech
```

```julia
# import packagu pro ploceni
using PyPlot
using PyCall
using LaTeXStrings
using Formatting

# latex fonty
rcParams = PyPlot.PyDict(PyPlot.matplotlib."rcParams")
rcParams["font.size"] = 14
rcParams["mathtext.fontset"] = "cm"
rcParams["mathtext.rm"] = "serif"
rcParams["text.usetex"] = "true";
```

```julia
my_dpi = 150
siz = 100
sizex = 16*siz
sizey = 9*siz

figure(figsize=(sizex/my_dpi, sizey/my_dpi), dpi=my_dpi)

x=gca()

line1 = x.plot(z_data*1e3, E_field_rad_data, color="blue", label=L" E_{0,z}^{\rm (rad)} (z) \approx a_1 z + a_3 z^3 + a_5 z^5")
line2 = x.plot(z_data*1e3, E_field_ax_data, color="red", label=L" E_{0,z}^{(6)} (z) \approx \sum_{0}^{8} a_k z^k")
x.set_xlabel(L"z \,\, \rm{[\rm{\mu} m]}", fontsize=16)
x.set_ylabel(L"E_{0,z} \, \,  \rm{[1/m]}", fontsize=16)

x.grid()
x.legend(framealpha=1, fontsize=16)

#savefig("/home/dan/diplomka_winfiles/Julia_vypocty_diplomka/axial_E_field/axial_E_field_witheq.svg", bbox_inches="tight")
```
